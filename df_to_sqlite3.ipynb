```python
import pandas as pd
import sqlite3
```

```python
final_data = "final_data.csv"
final_df = pd.read_csv(final_data)
```

```python
final_df = final_df.rename(columns={"Spring": "spring", "Summer": "summer", "Fall": "fall", "Winter": "winter", "Commodity": "commodity", "Commodity Type": "commodity_type", "Unnamed: 0": "id", "Date": "sale_date", "Weighted Avg Price": "weighted_avg_price"})
final_df.head()
```

```python
conn = sqlite3.connect('seasonal_produce.sqlite')
c = conn.cursor()
```

```python
c.execute('CREATE TABLE seasonal_produce (id, commodity, sale_date, weighted_avg_price, spring, summer, fall, winter, commodity_type)')
conn.commit()
```

```python
final_df.to_sql('seasonal_produce', conn, if_exists='replace', index = False)
```

```python
c.execute('''  
SELECT * FROM seasonal_produce
          ''')
for row in c.fetchall():
    print (row)
```
