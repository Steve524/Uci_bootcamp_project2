```python
# Import Dependencies
import pandas as pd
from pandas import DataFrame
import numpy as np
from sqlalchemy import create_engine
from config import user, password
```

```python
# Create a path to the csv and read it into a Pandas DataFrame
veg_path = "report_veg.csv"
fruit_path = "report_fruit.csv"

veg_df = pd.read_csv(veg_path, dtype= {'Commodity': str})
fruit_df = pd.read_csv(fruit_path, dtype= {'Commodity': str})

d_list = [veg_df, fruit_df]

merge_df = pd.concat(d_list)
yomp_df = merge_df.drop(columns=['Number of Stores','Organic', 'Environment', '% Marked Local', 'Low Price', 'High Price'])
clean_df = yomp_df.sort_values(['Date'])
clean_df
```

```python
#drop everything thats not in per pounds
drop_df = clean_df[clean_df.Unit.str.contains("per pound")]
drop_df
```
